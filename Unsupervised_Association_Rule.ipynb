#1.0 Running the Apriori Algorithm on the sample toy dataset

```python
#1. importing libraries
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

#2. Given toy_dataset
toy_dataset = [['Skirt', 'Sneakers', 'Scarf', 'Pants', 'Hat'],
               ['Sunglasses', 'Skirt', 'Sneakers', 'Pants', 'Hat'],
               ['Dress', 'Sandals', 'Scarf', 'Pants', 'Heels'],
               ['Dress', 'Necklace', 'Earrings', 'Scarf', 'Hat', 'Heels', 'Hat'],
               ['Earrings', 'Skirt', 'Skirt', 'Scarf', 'Shirt', 'Pants']]
toy_dataset
```

```python
#3. Convert the dataset to the required format using one-hot encoding
te = TransactionEncoder()
te_ary = te.fit(toy_dataset).transform(toy_dataset)
toy_transformed = pd.DataFrame(te_ary, columns=te.columns_)

toy_transformed
```

```python
#4. Run Apriori algorithm
frequent_itemsets = apriori(toy_transformed, min_support=0.5, use_colnames=True)

#5. Generate association rules
association_result = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)

#6. Display results
print("Frequent Itemsets:")
print(frequent_itemsets)

print("\nAssociation Rules:")
print(association_result)
```

```python
#6. Display result
#converting support result to a pandas dataframe
freq_items = pd.DataFrame(frequent_itemsets)
freq_items.sort_values(by = 'support', ascending = False).head(10)
```

```python
#6. Display Result
#converting association result to a pandas dataframe
association_result_df = pd.DataFrame(association_result)
association_result_df.sort_values(by = ['confidence', 'support', 'lift'], ascending = False).head(50)
```
